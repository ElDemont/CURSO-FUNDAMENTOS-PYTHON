```python
En una escuela se permite la inscripción de sus alumnos en las siguientes actividades deportivas: 
fútbol (f) y baloncesto (b). Realice un programa que permita ingresar el nombre de los estudiantes e inscribir a cada uno en uno o ambos deportes. 
La aplicación debe mostrar los siguientes listados:

Estudiantes inscritos en fútbol.

Estudiantes inscritos en baloncesto.

Estudiantes inscritos en ambos deportes.

Todos los estudiantes inscritos en algún deporte.

Estudiantes inscritos en solo un deporte.

Además, genere un menú con las opciones de la aplicación.
```

```python
def mostrar_estudiantes_futbol(futbol):
    print("\nEstudiantes inscritos en fútbol:")
    if futbol:
        for estudiante in futbol:
            print(estudiante)
    else:
        print("No hay estudiantes inscritos en fútbol.")

def mostrar_estudiantes_baloncesto(baloncesto):
    print("\nEstudiantes inscritos en baloncesto:")
    if baloncesto:
        for estudiante in baloncesto:
            print(estudiante)
    else:
        print("No hay estudiantes inscritos en baloncesto.")

def mostrar_estudiantes_ambos(futbol, baloncesto):
    print("\nEstudiantes inscritos en ambos deportes:")
    ambos = futbol.intersection(baloncesto)
    if ambos:
        for estudiante in ambos:
            print(estudiante)
    else:
        print("No hay estudiantes inscritos en ambos deportes.")

def mostrar_todos_estudiantes(futbol, baloncesto):
    print("\nTodos los estudiantes inscritos en algún deporte:")
    todos = futbol.union(baloncesto)
    if todos:
        for estudiante in todos:
            print(estudiante)
    else:
        print("No hay estudiantes inscritos en ningún deporte.")

def mostrar_estudiantes_solo_un_deporte(futbol, baloncesto):
    print("\nEstudiantes inscritos en solo un deporte:")
    solo_futbol = futbol - baloncesto
    solo_baloncesto = baloncesto - futbol
    if solo_futbol:
        for estudiante in solo_futbol:
            print(estudiante)
    if solo_baloncesto:
        for estudiante in solo_baloncesto:
            print(estudiante)
    if not solo_futbol and not solo_baloncesto:
        print("No hay estudiantes inscritos en solo un deporte.")

def menu():
    futbol = set()
    baloncesto = set()

    while True:
        print("\n--- Menú ---")
        print("1. Inscribir estudiante")
        print("2. Ver estudiantes inscritos en fútbol")
        print("3. Ver estudiantes inscritos en baloncesto")
        print("4. Ver estudiantes inscritos en ambos deportes")
        print("5. Ver todos los estudiantes inscritos en algún deporte")
        print("6. Ver estudiantes inscritos en solo un deporte")
        print("7. Salir")
        
        opcion = input("Seleccione una opción (1-7): ")
        
        if opcion == '1':
            nombre = input("Ingrese el nombre del estudiante: ")
            deportes = input("¿En qué deportes se inscribe (futbol/baloncesto/ambos)? (f/b/ambos): ").lower()
            if deportes == 'f' or deportes == 'ambos':
                futbol.add(nombre)
            if deportes == 'b' or deportes == 'ambos':
                baloncesto.add(nombre)
            print(f"{nombre} ha sido inscrito correctamente.")
        
        elif opcion == '2':
            mostrar_estudiantes_futbol(futbol)
        
        elif opcion == '3':
            mostrar_estudiantes_baloncesto(baloncesto)
        
        elif opcion == '4':
            mostrar_estudiantes_ambos(futbol, baloncesto)
        
        elif opcion == '5':
            mostrar_todos_estudiantes(futbol, baloncesto)
        
        elif opcion == '6':
            mostrar_estudiantes_solo_un_deporte(futbol, baloncesto)
        
        elif opcion == '7':
            print("¡Hasta luego!")
            break
        
        else:
            print("Opción no válida. Intente de nuevo.")

menu()
```

```python
realice un programa traductor de frutas del españoll al ingles. para ello, olicit el nombre de una fruta al usuario y muestre su traduccion al ingles
```

```python
frutas = {
    "manzana": "apple",
    "banana": "banana",
    "naranja": "orange",
    "pera": "pear",
    "uva": "grape",
    "piña": "pineapple",
    "fresa": "strawberry",
    "cereza": "cherry",
    "sandia": "watermelon",
    "mango": "mango"
}

print(" Frutas disponibles en español:")
for fruta in frutas.keys():
    print(f"- {fruta}")

fruta_espanol = input("\nIngrese el nombre de una fruta en español: ").lower()

traduccion = frutas.get(fruta_espanol)

if traduccion:
    print(f"La traducción de '{fruta_espanol}' al inglés es: {traduccion}")
else:
    print("Lo siento, la fruta ingresada no está en el diccionario.")
```
