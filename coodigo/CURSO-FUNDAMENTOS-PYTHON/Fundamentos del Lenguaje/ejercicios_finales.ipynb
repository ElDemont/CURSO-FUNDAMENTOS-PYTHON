```python
#trabajo 1
def calculadora(num1, num2, operador):
    try:
        if operador == "+":
            resultado = num1 + num2
        elif operador == "-":
            resultado = num1 - num2
        elif operador == "*":
            resultado = num1 * num2
        elif operador == "/":
            if num2 == 0:
                raise ZeroDivisionError("Error: No se puede dividir por cero.")
            resultado = num1 / num2
        elif operador == "**":
            resultado = num1 ** num2
        elif operador == "//":
            if num2 == 0:
                raise ZeroDivisionError("Error: No se puede dividir entre cero.")
            resultado = num1 // num2
        else:
            raise ValueError("Operador no válido.")
        
        print(f"Resultado: {num1} {operador} {num2} = {resultado}")
    
    except ValueError as e:
        print(f"Error: {e}")
    except ZeroDivisionError as e:
        print(f"{e}")

# Solicitar al usuario los valores
try:
    num1 = float(input("Ingrese el primer número: "))
    num2 = float(input("Ingrese el segundo número: "))
    operador = input("Ingrese el operador (+, -, *, /, **, //): ")
    calculadora(num1, num2, operador)
except ValueError:
    print("Por favor ingrese números válidos.")
```

```python
#trabajo 2

def agregar_producto(inventario, producto, cantidad):
    if producto in inventario:
        print(f"Error: El producto '{producto}' ya existe en el inventario.")
    else:
        inventario[producto] = cantidad
        print(f"Producto '{producto}' agregado con {cantidad} unidades.")

def actualizar_stock(inventario, producto, cantidad):
    if producto in inventario:
        inventario[producto] = cantidad
        print(f"Actualizando stock de '{producto}' a {cantidad}...")
    else:
        print(f"Error: El producto '{producto}' no existe en el inventario.")

def eliminar_producto(inventario, producto):
    if producto in inventario:
        del inventario[producto]
        print(f"Eliminando '{producto}'...")
    else:
        print(f"Error: El producto '{producto}' no existe en el inventario.")

def mostrar_inventario(inventario):
    print("Inventario actual:")
    for producto, cantidad in inventario.items():
        print(f"{producto}: {cantidad} unidades")

# Inventario inicial
inventario = {'manzanas': 50, 'naranjas': 30, 'peras': 20}
print(f"Inventario inicial: {inventario}")

# Actualizando, agregando y eliminando productos
actualizar_stock(inventario, 'peras', 25)
agregar_producto(inventario, 'bananas', 40)
eliminar_producto(inventario, 'naranjas')

# Inventario final
print(f"\nInventario final: {inventario}")
```

```python
#trabajo 3

import re

def analizar_texto(texto):
    if not texto:
        raise ValueError("Error: No se ha ingresado ningún texto.")
    
    # Convertir a minúsculas y separar las palabras
    texto = texto.lower()
    palabras = re.findall(r'\b\w+\b', texto)
    
    # Contar la frecuencia de cada palabra
    frecuencia = {}
    for palabra in palabras:
        if palabra in frecuencia:
            frecuencia[palabra] += 1
        else:
            frecuencia[palabra] = 1
            
    return frecuencia

# Solicitar al usuario un texto
try:
    texto = input("Ingrese un texto: ")
    resultado = analizar_texto(texto)
    
    print("Frecuencia de palabras:")
    for palabra, count in resultado.items():
        print(f"{palabra}: {count}")
except ValueError as e:
    print(e)
```

```python
#trabajo 4

def cm_a_pulgadas(cm):
    return cm / 2.54

def km_a_millas(km):
    return km * 0.621371

def convertir_unidad():
    print("Seleccione la conversión:")
    print("1. Centímetros a pulgadas")
    print("2. Kilómetros a millas")
    
    try:
        opcion = int(input("Opción: "))
        
        if opcion == 1:
            cm = float(input("Ingrese la cantidad en centímetros: "))
            resultado = cm_a_pulgadas(cm)
            print(f"{cm} cm = {resultado:.2f} pulgadas")
        elif opcion == 2:
            km = float(input("Ingrese la cantidad en kilómetros: "))
            resultado = km_a_millas(km)
            print(f"{km} km = {resultado:.2f} millas")
        else:
            print("Opción no válida.")
    except ValueError:
        print("Error: Por favor ingrese un número válido.")

# Llamada a la función principal
convertir_unidad()
```

```python
#trabajo 5

import random

def adivina_el_numero():
    numero_secreto = random.randint(1, 50)  # Número aleatorio entre 1 y 50
    intentos = 0
    adivinado = False
    
    print("¡Bienvenido al juego 'Adivina el Número'!")
    print("Estoy pensando en un número entre 1 y 50.")
    
    while not adivinado:
        try:
            intentos += 1
            numero_usuario = int(input(f"Intento {intentos} - Ingresa tu número: "))
            
            if numero_usuario < numero_secreto:
                print("El número secreto es mayor.")
            elif numero_usuario > numero_secreto:
                print("El número secreto es menor.")
            else:
                adivinado = True
                print(f"¡Felicidades! Has adivinado el número secreto: {numero_secreto}")
        
        except ValueError:
            print("Error: Por favor ingresa un número válido.")
    
# Iniciar el juego
adivina_el_numero()
```

```python
#trabajo 6

def calcular_factorial(n):
    if n < 0:
        raise ValueError("El número no puede ser negativo.")
    
    factorial = 1
    for i in range(1, n + 1):
        factorial *= i
    return factorial

def main():
    try:
        numero = int(input("Ingrese un número entero para calcular su factorial: "))
        resultado = calcular_factorial(numero)
        print(f"El factorial de {numero} es {resultado}")
    except ValueError as e:
        print(f"Error: {e}")

# Llamar a la función principal
main()
```

```python
#trabajo 7

from datetime import datetime

def dias_faltantes(fecha_str):
    try:
        fecha_usuario = datetime.strptime(fecha_str, "%d/%m/%Y")  # Convierte la cadena a un objeto datetime
        hoy = datetime.today()  # Obtiene la fecha actual
        dias_restantes = (fecha_usuario - hoy).days  # Calcula la diferencia de días
        return dias_restantes
    except ValueError:
        return "Error: El formato de la fecha no es válido. Use DD/MM/AAAA."

def main():
    fecha = input("Ingrese una fecha (DD/MM/AAAA): ")
    dias = dias_faltantes(fecha)
    
    if isinstance(dias, int):
        if dias > 0:
            print(f"Faltan {dias} días para el {fecha}.")
        elif dias == 0:
            print(f"Hoy es el día {fecha}.")
        else:
            print(f"La fecha {fecha} ya ha pasado {abs(dias)} días.")
    else:
        print(dias)

# Llamar a la función principal
main()
```

```python
#trabajo 8

def agregar_estudiante(registro, nombre, calificacion):
    if nombre in registro:
        print(f"Error: El estudiante {nombre} ya está registrado.")
    else:
        registro[nombre] = calificacion
        print(f"Estudiante {nombre} agregado con calificación {calificacion}.")

def actualizar_calificacion(registro, nombre, nueva_calificacion):
    if nombre not in registro:
        print(f"Error: El estudiante {nombre} no se encuentra en el registro.")
    else:
        registro[nombre] = nueva_calificacion
        print(f"Estudiante actualizado: {nombre} - {nueva_calificacion}")

def eliminar_estudiante(registro, nombre):
    if nombre not in registro:
        print(f"Error: El estudiante {nombre} no se encuentra en el registro.")
    else:
        del registro[nombre]
        print(f"Estudiante {nombre} eliminado del registro.")

def listar_estudiantes(registro):
    if not registro:
        print("No hay estudiantes registrados.")
    else:
        print("Lista de estudiantes y sus calificaciones:")
        for nombre, calificacion in registro.items():
            print(f"{nombre}: {calificacion}")

def menu():
    registro = {'Ana': 90, 'Luis': 78, 'Carlos': 85}  # Registro inicial
    while True:
        print("\nMenú:")
        print("1. Agregar estudiante")
        print("2. Actualizar calificación")
        print("3. Eliminar estudiante")
        print("4. Listar estudiantes")
        print("5. Salir")
        
        try:
            opcion = int(input("Opción: "))
            
            if opcion == 1:
                nombre = input("Ingrese el nombre del estudiante: ")
                calificacion = int(input("Ingrese la calificación del estudiante: "))
                agregar_estudiante(registro, nombre, calificacion)
            elif opcion == 2:
                nombre = input("Ingrese el nombre del estudiante: ")
                nueva_calificacion = int(input("Ingrese la nueva calificación: "))
                actualizar_calificacion(registro, nombre, nueva_calificacion)
            elif opcion == 3:
                nombre = input("Ingrese el nombre del estudiante a eliminar: ")
                eliminar_estudiante(registro, nombre)
            elif opcion == 4:
                listar_estudiantes(registro)
            elif opcion == 5:
                print("Saliendo del programa.")
                break
            else:
                print("Error: Opción no válida. Por favor, seleccione una opción del menú.")
        except ValueError:
            print("Error: Por favor ingrese un número entero para seleccionar una opción.")

# Iniciar el menú
menu()
```

```python
#trabajo 9

import random
import string

def generar_contrasena(longitud):
    caracteres = string.ascii_letters + string.digits + string.punctuation
    contrasena = ''.join(random.choice(caracteres) for _ in range(longitud))
    return contrasena

def main():
    try:
        longitud = int(input("Ingrese la longitud deseada para la contraseña: "))
        if longitud <= 0:
            raise ValueError("La longitud debe ser un número entero positivo.")
        contrasena = generar_contrasena(longitud)
        print(f"Contraseña generada: {contrasena}")
    except ValueError as e:
        print(f"Error: {e}")

main()
```

```python
#trabajo 10

import os

def buscar_archivos(directorio, subcadena):
    try:
        archivos = os.listdir(directorio)
        encontrados = [archivo for archivo in archivos if subcadena in archivo]
        
        if encontrados:
            print(f'\nArchivos encontrados que contienen "{subcadena}":')
            for archivo in encontrados:
                print(archivo)
        else:
            print(f'\nNo se encontraron archivos que contengan "{subcadena}".')
    
    except FileNotFoundError:
        print("Error: El directorio no existe.")
    except PermissionError:
        print("Error: Permisos insuficientes para acceder al directorio.")
    except Exception as e:
        print(f"Ocurrió un error inesperado: {e}")

def main():
    ruta = input("Ingrese la ruta del directorio: ")
    subcadena = input("Ingrese la subcadena a buscar: ")
    buscar_archivos(ruta, subcadena)

main()
```
